```python
# فرض بر این است که مسیر path از kagglehub گرفته شده است و فایل 'Last.fm_data.csv' آنجاست

import os
import pandas as pd

csv_path = os.path.join(path, 'Last.fm_data.csv')
df = pd.read_csv(csv_path)
user_artist_counts = df.groupby(['Username', 'Artist']).size().reset_index(name='plays')
user_item = user_artist_counts.pivot_table(index='Username', columns='Artist', values='plays', fill_value=0)

from sklearn.metrics.pairwise import cosine_similarity

item_sim = pd.DataFrame(
    cosine_similarity(user_item.T),
    index=user_item.columns,
    columns=user_item.columns
)
user_sim = pd.DataFrame(
    cosine_similarity(user_item),
    index=user_item.index,
    columns=user_item.index
)
artist_meta = {artist: {'artist': artist} for artist in user_item.columns}

def item_based_recommend(user_id, N=5):
    if user_id not in user_item.index: return []
    listened = user_item.loc[user_id]
    listened = listened[listened > 0]
    scores = pd.Series(dtype='float64')
    explanations = {}
    for artist in listened.index:
        sim_scores = item_sim[artist] * listened[artist]
        sim_scores = sim_scores.drop(listened.index, errors='ignore')
        for s in sim_scores.index:
            if s not in scores:
                scores[s] = 0
                explanations[s] = artist
            scores[s] += sim_scores[s]
    top_artists = scores.sort_values(ascending=False).head(N)
    results = []
    for artist in top_artists.index:
        meta = artist_meta.get(artist, {})
        expl = f"چون به هنرمند '{explanations[artist]}' زیاد گوش دادید"
        results.append({'artist': artist, 'score': float(top_artists[artist]), 'reason': expl, **meta})
    return results

def user_based_recommend(user_id, N=5):
    if user_id not in user_item.index: return []
    similar_users = user_sim.loc[user_id].sort_values(ascending=False).iloc[1:4].index
    artist_scores = {}
    reasons = {}
    for u in similar_users:
        for artist in user_item.loc[u][user_item.loc[u]>0].index:
            if user_item.loc[user_id][artist]==0:
                artist_scores[artist] = artist_scores.get(artist, 0) + user_item.loc[u][artist]
                reasons[artist] = f"چون کاربر مشابه شما ({u}) این هنرمند را گوش داده"
    recs = sorted(artist_scores.items(), key=lambda x: -x[1])[:N]
    return [{'artist': s, 'score': float(sc), 'reason': reasons[s], **artist_meta.get(s, {})} for s,sc in recs]

def smart_playlist(seed_artist, N=5):
    if seed_artist not in item_sim.index: return []
    sim_scores = item_sim[seed_artist].sort_values(ascending=False)
    recs = [s for s in sim_scores.index if s!=seed_artist][:N]
    return [
        {'artist': s, **artist_meta.get(s, {}), 'reason': f"شباهت زیاد به {seed_artist}"}
        for s in recs
    ]

from flask import Flask, jsonify
from pyngrok import ngrok
import threading

app = Flask(__name__)

@app.route('/')
def home():
    return '''
    <html>
    <head>
      <meta charset="utf-8">
      <title>Music Recommender Demo</title>
      <style>
        body { font-family: sans-serif; direction: rtl; margin:2em; }
        .rec { border:1px solid #ccc; border-radius:7px; margin:1em 0; padding:1em; }
      </style>
    </head>
    <body>
      <h2>دموی سیستم توصیه‌گر موسیقی</h2>
      <p>بر پایه دیتاست Last.fm واقعی.<br>
      مقدار user همان ستون Username و مقدار artist همان ستون Artist در دیتاست است.</p>
      <div>
  <b>توصیه هنرمند برای کاربر:</b><br>
  <input id="uid" placeholder="مثلاً Babs_05" style="width:12em">
  <button onclick="getRec()">توصیه آیتم‌بیس</button>
  <button onclick="getRecUser()">توصیه یوزر-بیس</button>
</div>
<div>
  <b>پلی‌لیست هنرمند مشابه:</b><br>
  <input id="seed" placeholder="مثلاً !!!" style="width:12em">
  <button onclick="getPlaylist()">پلی‌لیست</button>
</div>

      <div id="result"></div>
      <script>
        function showRecs(data) {
          let html = "";
          if(!data || data.length==0){ html="<div>پیشنهادی یافت نشد.</div>"; }
          else data.forEach(r=>{
            html+=`<div class="rec">
              <b>هنرمند:</b> ${r.artist||'-'} <br>
              <b>امتیاز:</b> ${r.score!==undefined ? r.score.toFixed(2) : "-"}<br>
              <b>دلیل:</b> ${r.reason}
            </div>`;
          });
          document.getElementById('result').innerHTML=html;
        }
        function getRec(){
          let uid=document.getElementById('uid').value;
          fetch('/recommend/item/'+uid).then(r=>r.json()).then(showRecs);
        }
        function getRecUser(){
          let uid=document.getElementById('uid').value;
          fetch('/recommend/user/'+uid).then(r=>r.json()).then(showRecs);
        }
        function getPlaylist(){
          let sid=document.getElementById('seed').value;
          fetch('/playlist/'+sid).then(r=>r.json()).then(showRecs);
        }
      </script>
    </body>
    </html>
    '''

@app.route('/recommend/item/<user_id>')
def rec_item(user_id):
    return jsonify(item_based_recommend(user_id))

@app.route('/recommend/user/<user_id>')
def rec_user(user_id):
    return jsonify(user_based_recommend(user_id))

@app.route('/playlist/<seed_artist>')
def playlist(seed_artist):
    return jsonify(smart_playlist(seed_artist, N=5))

def run_app():
    app.run(port=5000)

thread = threading.Thread(target=run_app)
thread.start()

public_url = ngrok.connect(5000)
print('Your app is live at:', public_url)
```
